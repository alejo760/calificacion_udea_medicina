```python
df_def.set_index('name', inplace=True)
```

```python
#get a list of unique values in name column
names = df_def.index.unique()
#change the type od names to a list
names = names.tolist()
names
```

```python
#if the count of value of 'nucleo' == 'Rotación Hospitalaria' for each 'name' is < 5 create a new row with the same 'name'  and value 'Rotación Hospitalaria' in column 'nucleo' and 'score' == 0
for name in names:
        count = df_def.loc[df_def['name'] == name, 'nucleo'].value_counts().get('Rotación Hospitalaria', 0)
        if count < 5:
            # Create a new row with the same 'name' and value 'Rotación Hospitalaria' in column 'nucleo'
            new_row = pd.DataFrame({'name': [name], 'nucleo': ['Rotación Hospitalaria'], 'score': [0]})
            df_def = pd.concat([df_def, new_row], ignore_index=True)
   
    
```

```python
#droput rows with column 'score' == Nan
df_def = df_def.dropna(subset=['score'])
#change the type of column 'score' to float
df_def['score'] = df_def['score'].astype(float)

```
