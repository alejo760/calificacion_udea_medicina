```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_excel('Historial de notas.xlsx')
df1.head()
```

```python
#Select the rows that have == "Total del curso" in the column Ítem de calificación and save it in a new dataframe
df2=df1[df1['Ítem de calificación']=='Total del curso']
df2.head()
```

```python
#Select the rows that have == "aggregation" in the column 'Fuente'  and save it in a new dataframe
df3=df2[df2['Fuente']=='aggregation']
df3.head()
```

```python
#type to float the column 'Calificación revisada'
df3['Calificación revisada']=df3['Calificación revisada'].astype(float)
```

```python
#divide the column 'Calificación revisada' by 100  then multiply by 5 and save it in a new column 'Calificación revisada escala 0 a 5'
df3['Calificación revisada escala 0 a 5']=df3['Calificación revisada']/100*5
df3.head()
```

```python
mean=df3['Calificación revisada escala 0 a 5'].mean()
print(mean)
```

```python
#eliminate the repeated rows in the column 'Dirección de correo' keeping the max value of the column 'Calificación revisada escala 0 a 5'
df4=df3.groupby('Dirección de correo').max()
df4.head()
```

```python
#download the dataframe as a excel file
df4.to_excel('Notas_curso EKG.xlsx',index=False)
```
